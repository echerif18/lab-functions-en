## 1. Write a function that returns the greater of two numbers

```python
def greater(a: int,b:int) -> int:
    # your code here
    """
    This function returns the greater value of two numbers.

    Args:
        a (int): The first number.
        b (int): The second number.

    Returns:
        int: The largest element of two numbers.
    """
    if a>b:
        return a
    else:
        return b
```

## 2. Now write a function that returns the largest element on a list

```python
def greatest(ls:list) -> any:
    # your code here
    """
    This function returns the largest element on a list.

    Args:
        ls (list)

    Returns:
        any: The largest element.
    """
    greatest = ls[0]
    for i in ls[1:]:
        if i > greatest:
            greatest = i
    return greatest
```

```python
# This will test your function 
test_greatest(greatest)
```

## 3. Write a function that sums all the elements on a list

```python
def sum_all(lst:list) -> any:
    # your code here
    """
    This function returns the summation of elements on a list.

    Args:
        lst (list): list of numbers >> add excemptionis a plus!

    Returns:
        any: The sum of the elements.
    """
    sum = lst[0]
    for i in lst[1:]:
        sum+=i
    return sum
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n: int) -> int:
    # your code here
    """
    This function returns the factorial of a number.

    Args:
        n (int)

    Returns:
        any: Return the result of the factorial
    """
    if n <= 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ðŸ¤”`

```python
def unique(lst_un: list) -> list:
    # your code here
    """
    This function returns the unique elements from a list.

    Args:
        n (int)

    Returns:
        list: Returns a list of the unique elements
    """
    ls_uni = []
    for i in lst_un:
        if i not in ls_uni:
            ls_uni.append(i)
    return ls_uni
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string:str) -> bool:
    # your code here
    """
    This function checks if a string is a pangram

    Args:
        string (str)

    Returns:
        bool: Returns whether a string is a pangram.
    """
    alphabets = [chr(i) for i in range(ord('a'), ord('z') + 1)]

    word_ = string.lower()

    for i in alphabets:
        if i not in word_:
            return False
    
    return True
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password:str) -> bool:
    # your code here
    """
    This function checks if a password is strong

    Args:
        string (str)

    Returns: 
        bool: Returns whether a password is strong
    """
    
    if(len(password)<8):
        return False
    elif(not any(i.islower() for i in password)):
        return False
    elif(not any(i.isupper() for i in password)):
        return False
    elif(not any(i.isnumeric() for i in password) ):
        return False
    elif(not(any(not (i.isalnum()) for i in password))):
        return False
    else:
        return True
```

```python
# This will test your function 
test_pass(check_pass)
```
